```
install.packages("ggplot2")  ##安裝套件
install.packages("SportsAnalytics")
```

```
library(SportsAnalytics)   ##載入套件
library(ggplot2)
```

```
NBA1516<-fetch_NBAPlayerStatistics("15-16") #讀入資料
```

```
#製作散佈圖 qplot(x軸名稱,y軸名稱,data=使用資料)
qplot(FieldGoalsAttempted,TotalPoints,data=NBA1516,color=Position)
```

```
#製作長條圖 geom_bar()
ggplot()+geom_bar(data=NBA1516,aes(x=FieldGoalsAttempted,y=TotalPoints),stat="identity") #stat="identity"直接畫數字
```

```
df<-data.frame(GamesPlayed=factor(rep(letters[1:3],each=10)),y=rnorm(30))
```

```
#製作盒狀圖 geom_boxplot()
ggplot(df,aes(x=GamesPlayed,y=y))+geom_point()+geom_boxplot()
```

```
df$z<-df$y+rnorm(30)
```

```
#製作折線圖 geom_line()
ggplot(df,aes(x=z,y=y))+geom_line()
```
