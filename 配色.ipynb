```
#安裝套件
install.packages("SportsAnalytics")
```

```
#載入套件
library(SportsAnalytics)
library(ggplot2)
```

```
#讀入資料
NBA1819 <- fetch_NBAPlayerStatistics("18-19")
```

```
#利用order()函數由大到小排序(decreasing = T)
NBA1819Order<-NBA1819[order(NBA1819$FieldGoalsMade,decreasing =T),]
NBA1819Order[1:20,"Name"] #逗號前方取1-20列；逗號後方放"Name"，表示取名稱為Name欄位
```

```
#用subset()做欄位條件搜尋
subset(NBA1819,Position=="SG")
```

```
#用grepl()做字串條件搜尋
NBA1819[grepl("Booker",NBA1819$Name),]
```

```
install.packages("devtools")
devtools::install_github('cttobin/ggthemr')
library(ggthemr)
ggthemr=("dust")
```

```
#製作散佈圖 qplot(x軸名稱,y軸名稱,data=使用資料)
qplot(GamesPlayed,FieldGoalsMade,data=NBA1819)
```

```
NBA1819$GamesPlayed <- as.factor(NBA1819$GamesPlayed)
```

```
#畫盒狀圖
ggplot(NBA1819,aes(GamesPlayed,FieldGoalsMade))+geom_boxplot()
```

```
#畫長條圖
ggplot()+geom_bar(data=NBA1819,aes(x=GamesPlayed,y=),stat="identity")
```

```
#畫geom_line()
ggplot(NBA1819,aes(GamesPlayed,FieldGoalsMade))+geom_line()
```
